## 1. Configuración e Importación de Librerías

Cargaremos las herramientas gráficas standard en ciencia de datos (`matplotlib`, `seaborn`) y manipulación (`pandas`, `numpy`).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = [10, 6]
print("Librerías gráficas listas.")
```

```python
# Cargar datos limpios de la ETL
df = pd.read_csv('data/prestamos_campo_limpio.csv')
print(f"Filas cargadas: {df.shape[0]}")
df.head()
```

--- 
## 2. Análisis Estadístico y Distribución del Target

Revisamos cómo se distribuyen las tasas de interés asignadas en el histórico (`Tasa_Interes_EA`).

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Tasa_Interes_EA'], kde=True, color='#15803d', bins=25)
plt.title('Distribución de Tasas de Interés Efectivas Anuales (E.A. %)', fontsize=13, fontweight='bold')
plt.xlabel('Tasa E.A. %')
plt.ylabel('Frecuencia')
plt.show()
```

--- 
## 3. Relación de Atributos del Campo contra la Tasa

### 3.1 Impacto de los Factores de Fomento (Subsidios y Garantías)
Graficaremos el impacto del subsidio LEC de Finagro (`Subsidio_Gobierno`) y el aval del Fondo Agropecuario de Garantías (`Garantia_Respaldada`) en las tasas financieras.

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Subsidio LEC vs Tasa
sns.boxplot(data=df, x='Subsidio_Gobierno', y='Tasa_Interes_EA', ax=axes[0], palette='YlOrBr')
axes[0].set_title('Subsidio LEC Finagro vs Tasa E.A. %', fontsize=12, fontweight='bold')
axes[0].set_xticklabels(['Sin Subsidio', 'Con Subsidio LEC'])
axes[0].set_xlabel('')

# Garantía FAG vs Tasa
sns.boxplot(data=df, x='Garantia_Respaldada', y='Tasa_Interes_EA', ax=axes[1], palette='Greens')
axes[1].set_title('Respaldo Garantía FAG vs Tasa E.A. %', fontsize=12, fontweight='bold')
axes[1].set_xticklabels(['Sin Garantía Real', 'Con Garantía FAG'])
axes[1].set_xlabel('')

plt.tight_layout()
plt.show()
```

### Interpretación:
- **Subsidio de Gobierno (Finagro LEC):** Hay un desplazamiento drástico hacia abajo de la mediana de las tasas cuando existe subsidio, disminuyendo alrededor de 5.5 puntos porcentuales.
- **Garantía Respaldo FAG:** Las propiedades o productores respaldados por garantías del Fondo FAG cotizan a tasas notablemente más bajas, disminuyendo cerca de 3.2 puntos.

### 3.2 Relación de Experiencia y Plazo con la Tasa

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Experiencia vs Tasa
sns.regplot(data=df, x='Experiencia_Anios', y='Tasa_Interes_EA', ax=axes[0], 
            scatter_kws={'alpha':0.3, 'color':'#16a34a'}, line_kws={'color':'#dc2626'})
axes[0].set_title('Años de Experiencia Agrícola vs Tasa E.A. %', fontsize=12, fontweight='bold')
axes[0].set_xlabel('Años de Experiencia')
axes[0].set_ylabel('Tasa de Interés E.A. %')

# Plazo vs Tasa
sns.boxplot(data=df, x='Plazo_Meses', y='Tasa_Interes_EA', ax=axes[1], palette='Purples')
axes[1].set_title('Plazo del Préstamo (Meses) vs Tasa E.A. %', fontsize=12, fontweight='bold')
axes[1].set_xlabel('Plazo (Meses)')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

--- 
## 4. Matriz de Correlación de Pearson

Evaluamos la correlación numérica global entre las variables.

```python
columnas_interes = ['Monto_Prestamo_COP', 'Plazo_Meses', 'Edad_Agricultor', 'Experiencia_Anios', 'Ingresos_Mensuales_COP', 'Garantia_Respaldada', 'Subsidio_Gobierno', 'Tasa_Interes_EA']
corr_matrix = df[columnas_interes].corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5, vmin=-1, vmax=1)
plt.title('Matriz de Correlación de Pearson - Crédito Agropecuario', fontsize=13, fontweight='bold', pad=15)
plt.show()
```

--- 
## 5. Conclusiones del EDA

1. **Impacto Fuerte del Subsidio (r ≈ -0.73):** El Subsidio Finagro LEC representa la influencia negativa más fuerte sobre la tasa, lo que concuerda con las políticas estatales de fomento financiero rural.
2. **Respaldo FAG (r ≈ -0.42):** La presencia de garantías FAG mitiga severamente el riesgo financiero de la entidad bancaria, disminuyendo la tasa.
3. **Inexistencia de Multicolinealidad Extrema:** Ningún par de variables predictoras presenta correlaciones que superen r = 0.35, lo que garantiza supuestos óptimos para ajustar un modelo de **Regresión Lineal Múltiple** en el **Cuaderno 3**.
