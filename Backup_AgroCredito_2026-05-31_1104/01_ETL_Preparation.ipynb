## 1. Identificación del Problema (Business Understanding)

### Contexto y Problema de Negocio
El sector rural colombiano requiere capital para tecnificar la producción, adquirir maquinaria y financiar insumos. Sin embargo, las tasas de interés del mercado tradicional a menudo no reflejan adecuadamente factores de mitigación de riesgo como las garantías del Fondo Agropecuario de Garantías (FAG), años de experiencia del agricultor o subsidios del gobierno a través de Finagro (Líneas Especiales de Crédito - LEC).

### Objetivo del Modelo de ML
Desarrollar un modelo de **Regresión Lineal** que prediga de manera personalizada la **Tasa de Interés Efectiva Anual (Tasa_Interes_EA)** para préstamos del campo colombiano, considerando perfiles financieros y factores de fomento nacional. Esto permitirá automatizar la tasación transparente de créditos agrícolas.

## 3. Preparación de Datos (Data Preparation - ETL)

Aplicaremos procesos de ETL para limpiar el dataset:
1. **Identificar nulos**.
2. **Imputar nulos** con técnicas estadísticas óptimas.
3. **Dar formato correcto** a los tipos de datos.

```python
# Revisar nulos
nulos = df_raw.isnull().sum()
print("Valores nulos encontrados:")
print(nulos[nulos > 0])
```

### Estrategia de Imputación:
- **Experiencia_Anios:** Se rellenará con la **mediana** para proteger ante posibles agricultores centenarios o atípicos.
- **Garantia_Respaldada:** Al ser binaria (0 o 1), se rellenará con la **moda** (el valor más común).

```python
df_clean = df_raw.copy()

# Imputar experiencia
mediana_exp = df_clean['Experiencia_Anios'].median()
df_clean['Experiencia_Anios'] = df_clean['Experiencia_Anios'].fillna(mediana_exp)

# Imputar garantía
moda_gar = df_clean['Garantia_Respaldada'].mode()[0]
df_clean['Garantia_Respaldada'] = df_clean['Garantia_Respaldada'].fillna(moda_gar)

assert df_clean.isnull().sum().sum() == 0
print("Nulos imputados correctamente.")
```

```python
# Conversión de tipos de datos a enteros óptimos
df_clean['Plazo_Meses'] = df_clean['Plazo_Meses'].astype(int)
df_clean['Edad_Agricultor'] = df_clean['Edad_Agricultor'].astype(int)
df_clean['Experiencia_Anios'] = df_clean['Experiencia_Anios'].astype(int)
df_clean['Garantia_Respaldada'] = df_clean['Garantia_Respaldada'].astype(int)
df_clean['Subsidio_Gobierno'] = df_clean['Subsidio_Gobierno'].astype(int)

df_clean.dtypes
```

## 4. Exportación del Dataset Limpio

El dataset limpio está listo. Se exporta como `data/prestamos_campo_limpio.csv` para las fases de análisis EDA y modelado.

```python
output_file = 'data/prestamos_campo_limpio.csv'
df_clean.to_csv(output_file, index=False)
print(f"Dataset guardado con éxito en: {output_file}")
```
